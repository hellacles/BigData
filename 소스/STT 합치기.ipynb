# STT 에서 추출된 데이터프레임을 합쳐봅시다

```python
# 데이터프레임 불러오기
import pandas as pd

df1 = pd.read_excel('./data/text/train_data_re.xlsx')
df2 = pd.read_excel('./data/text/박성준녹음.xlsx')
df2.drop(df2.columns[0],axis=1,inplace=True)
df3 = pd.read_excel('./data/text/연하녹음.xlsx')
df3.drop(df3.columns[0],axis=1,inplace=True)
df4 = pd.read_excel('./data/text/재환이녹음.xlsx')
df4.drop(df4.columns[0],axis=1,inplace=True)
df5 = pd.read_excel('./data/text/클라우드녹음.xlsx')
df5.drop(df5.columns[0],axis=1,inplace=True)

display(df1.tail())
display(df2.tail())
display(df3.tail())
display(df4.tail())
display(df5.tail())
```

```python
# 데이터프레임을 합쳐보자

all_of_data = pd.concat([df1, df2, df3, df4, df5])
display(all_of_data)

all_of_data.to_excel('./data/text/최종.xlsx')
```
